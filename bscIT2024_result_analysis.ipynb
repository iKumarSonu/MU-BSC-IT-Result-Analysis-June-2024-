# Mumbai University Bsc IT Result analysis (June 2024)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("updated_results.csv")
```

```python
# Convert 'cgpi' column to numeric (automatically handles '--' as NaN)
df['CGPI'] = pd.to_numeric(df['CGPI'], errors='coerce')

# Calculate the mean ignoring NaN values
average_cgpi = df['CGPI'].mean()

print(f"Average CGPI: {average_cgpi}")
```

# TOP COLLEGES

```python
top_colleges = df.groupby('College_name')['CGPI'].mean().sort_values(ascending=False)
```

```python
top_colleges
```

# TOP STUDENTS

```python
top_students = df[['Name', 'CGPI']].sort_values(by='CGPI', ascending=False)
top_students
```

# Performance Category

```python
def performance_category(row):
    if row['CGPI'] >= 8.0:
        return 'Excellent'
    elif row['CGPI'] >= 6.0:
        return 'Good'
    elif row['CGPI'] >= 4.0:
        return 'Average'
    elif row['CGPI'] < 4.0 and row['CGPI'] >= 0:
        return 'Poor'
    else:
        return 'Unknown'

df['Performance_category'] = df.apply(performance_category, axis=1)
df
```

## Performance Category Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['Performance_category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Performance Category Distribution')
plt.ylabel('')
plt.show()
```

# Correlation Analysis

```python
df[['SEM-I', 'SEM-II', 'SEM-III', 'SEM-IV', 'SEM-V', 'SEM-VI', 'CGPI']] = df[['SEM-I', 'SEM-II', 'SEM-III', 'SEM-IV', 'SEM-V', 'SEM-VI', 'CGPI']].apply(pd.to_numeric, errors='coerce')

correlation_matrix = df[['SEM-I', 'SEM-II', 'SEM-III', 'SEM-IV', 'SEM-V', 'SEM-VI', 'CGPI']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Pass / Fail Analysis

```python
value_counts = df['PASS/FAIL'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
bars = value_counts.plot(kind='bar', color=['green', 'red', 'grey'])  # Custom colors for Pass/Fail
plt.title('Pass/Fail Distribution')
plt.xlabel('Pass/Fail Status')
plt.ylabel('Count')

# Adding counts on top of bars
for bar in bars.patches:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, int(height), 
             ha='center', va='bottom')

plt.xticks(ticks=[i for i in range(len(value_counts.index))],
           labels=[label for label in value_counts.index], rotation=0)
plt.show()
```

# Percentage of students passed

```python
total_students = len(df)

# Count the number of students who passed
passed_students = df['PASS/FAIL'].value_counts().get('Pass', 0)

# Calculate the percentage of students who passed
percentage_passed = (passed_students / total_students) * 100

print(f"Percentage of students who passed: {percentage_passed:.2f}%")
```

# Time Series Analysis
## Preparing the Data

```python
# Convert all columns to numeric, forcing errors to NaN
columns_to_extract = ['Seat_no', 'SEM-I', 'SEM-II', 'SEM-III', 'SEM-IV', 'SEM-V', 'SEM-VI']
df = df[columns_to_extract]
df = df.apply(pd.to_numeric, errors='coerce')

# Check data types again
print(df.dtypes)
```

```python
# Reshape the DataFrame
df_melted = df.melt(id_vars=['Seat_no'], var_name='Semester', value_name='Score')

# Handle NaN values
df_melted['Score'] = df_melted['Score'].fillna(0)  # Or use dropna() based on your needs

# Extract semester numbers
df_melted['Semester_Num'] = df_melted['Semester'].str.extract('(\d+)', expand=False).astype(float)
df_melted['Semester_Num'] = df_melted['Semester_Num'].fillna(0).astype(int)

# Sort by student and semester number
df_melted = df_melted.sort_values(by=['Seat_no', 'Semester_Num'])

# Calculate average score per semester
average_scores = df_melted.groupby('Semester')['Score'].mean()
```

## Trend Analysis

```python
# Plotting
plt.figure(figsize=(10, 6))
average_scores.plot(kind='line', marker='o')
plt.title('Average Scores Over Semesters')
plt.xlabel('Semester')
plt.ylabel('Average Score')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

### After Time Series Analysis, the line graph above shows a decreasing trend of the results in Students result in semesters from Sem 1 to Sem 6. This can be because the empty value were replaced by 0 to get the mean of scores.
